# 1. Python Data Types

Python provides several built-in data types. The four fundamental data types are:

- Integer (int)
- Float (float)
- Boolean (bool)
- String (str)

Each data type stores different kinds of information.

```python
# Integer example
age = 20

# Float example
height = 5.7

# Boolean example
is_student = True

# String example
name = "Noor"

print(age)
print(height)
print(is_student)
print(name)
```

## Type Checking

Python provides the `type()` function to determine the data type of a variable.

```python
print(type(age))
print(type(height))
print(type(is_student))
print(type(name))
```

# 2. Type Conversion

Python allows converting one data type into another using built-in functions.

Common conversion functions:

- int()
- float()
- str()
- bool()

```python
# Integer to String
age_string = str(age)

# String to Float
price = "99.99"
price_float = float(price)

# Float to Integer
height_integer = int(height)

print(age_string)
print(type(age_string))

print(price_float)
print(type(price_float))

print(height_integer)
print(type(height_integer))
```

# 3. Variables

A variable is a named container that stores data.

Good variable names follow the PEP 8 naming convention.

Examples:

- first_name
- student_age
- total_marks

Avoid names like:

- x
- y
- abc

```python
student_name = "Noor"
student_age = 20
total_marks = 95

print(student_name)
print(student_age)
print(total_marks)
```

# 4. Input and Output

Python uses:

- input() → to receive input from the user.
- print() → to display output.

```python
name = input("Enter your name: ")

print(f"Hello: {name}")
```

# 5. Interactive Calculator

```python
"""
This program performs basic arithmetic operations
using user input.
"""

try:
    # Get input from user
    first_number = float(input("Enter the first number: "))
    second_number = float(input("Enter the second number: "))

    # Perform calculations
    sum_result = first_number + second_number
    difference_result = first_number - second_number
    product_result = first_number * second_number

    print("\n" + "=" * 40)
    print("CALCULATOR RESULTS")
    print("=" * 40)

    print(f"The sum of {first_number} and {second_number} is: {sum_result}")
    print(f"The difference of {first_number} and {second_number} is: {difference_result}")
    print(f"The product of {first_number} and {second_number} is: {product_result}")

    if second_number != 0:
        quotient_result = first_number / second_number
        print(f"The quotient of {first_number} and {second_number} is: {quotient_result}")
    else:
        print("Division by zero is not allowed.")

except ValueError:
    print("Invalid input! Please enter only numeric values.")
```

# 8. Dynamic Typing

Python is dynamically typed.

A variable can hold different data types during program execution.

```python
value = 100
print(value)
print(type(value))

value = "Python"
print(value)
print(type(value))
```
