## Program 1 - Number Classifier

````python
# Number Classifier Program

number = int(input("Enter a number: "))

if number > 0:
    print("Number is positive")
elif number < 0:
    print("Number is negative")
else:
    print("Number is zero")
````

### Program Logic

This program takes a number from the user and determines whether it is:

- Positive
- Negative
- Zero

It uses `if`, `elif`, and `else` statements.

## Program 2 - Age Verification System

````python
print("=" * 40)
print("Age Verification System")
print("=" * 40)

try:
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))

    if age < 0:
        print("Invalid age! Age cannot be negative.")

    elif age < 13:
        print(f"Hello {name}! You are a Child.")
        print("Enjoy learning and having fun!")

    elif age < 18:
        print(f"Hello {name}! You are a Teenager.")
        print("You have many opportunities ahead.")

    elif age < 65:
        print(f"Hello {name}! You are an Adult.")
        print("Wishing you success in your journey!")

    else:
        print(f"Hello {name}! You are a Senior Citizen.")
        print("Stay healthy and enjoy life!")

except ValueError:
    print("Invalid input! Please enter a valid numeric age.")
````

### Program Logic

This program:

- Takes the user's name and age.
- Validates the age.
- Classifies the user into:
  - Child
  - Teenager
  - Adult
  - Senior Citizen
- Handles invalid input using `try-except`.

## Program 3 - Grade Calculator

````python
print("=" * 40)
print("Grade Calculator")
print("=" * 40)

try:
    marks = float(input("Enter your marks (0-100): "))

    if marks < 0 or marks > 100:
        print("Invalid marks! Please enter marks between 0 and 100.")

    elif marks >= 90:
        grade = "A"
        message = "Excellent work!"

    elif marks >= 80:
        grade = "B"
        message = "Good job!"

    elif marks >= 70:
        grade = "C"
        message = "Nice effort!"

    elif marks >= 60:
        grade = "D"
        message = "Keep improving!"

    else:
        grade = "F"
        message = "Don't give up. Practice more!"

    if 0 <= marks <= 100:
        print(f"\nMarks : {marks}")
        print(f"Grade : {grade}")
        print(f"Feedback : {message}")

except ValueError:
    print("Invalid input! Please enter numeric marks only.")
````

### Program Logic

This program:

- Accepts marks between 0 and 100.
- Assigns a letter grade.
- Displays feedback.
- Handles invalid and non-numeric input.

## Program 4 - Nested If Practice

````python
print("=" * 40)
print("Nested If Practice")
print("=" * 40)

try:
    number = int(input("Enter a number: "))

    if number > 0:
        print("The number is positive.")

        if number % 2 == 0:
            print("It is an Even number.")
        else:
            print("It is an Odd number.")

    elif number == 0:
        print("The number is Zero.")

    else:
        print("The number is Negative.")

except ValueError:
    print("Invalid input! Please enter a valid integer.")
````

### Program Logic

This program demonstrates a nested `if` statement.

If the number is positive, another `if` statement checks whether it is even or odd.

Otherwise, it checks whether the number is zero or negative.

# Practical Use Cases of Conditional Statements

Conditional statements are widely used in real-world programming to make decisions based on different conditions.

## 1. User Authentication

Conditional statements verify a user's credentials before granting access.

**Example:**

```python
if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Username or Password")
```

---

## 2. Data Validation

Conditionals check whether the user's input is valid before processing it.

**Example:**

```python
age = int(input("Enter your age: "))

if age >= 0:
    print("Valid age")
else:
    print("Invalid age")
```

---

## 3. Program Flow Control

Conditional statements determine which part of the program should execute based on a condition.

**Example:**

```python
marks = 75

if marks >= 50:
    print("Pass")
else:
    print("Fail")
```

# Flow Control in Python

Flow control is the order in which a program executes instructions.

The `if`, `elif`, and `else` statements control the execution path of a program.

- `if` checks the first condition.
- If the condition is `True`, its block executes.
- If the condition is `False`, Python checks the `elif` condition.
- If none of the conditions are `True`, the `else` block executes.

### Example

```python
temperature = 35

if temperature >= 40:
    print("Very Hot")
elif temperature >= 30:
    print("Hot")
else:
    print("Normal")
```

This allows the program to make decisions and execute different blocks of code based on different situations.

# Common Pitfalls in Conditional Statements

## 1. Assignment (`=`) vs Comparison (`==`)

The assignment operator (`=`) stores a value in a variable, while the comparison operator (`==`) checks whether two values are equal.

**Incorrect**

```python
if age = 18:
    print("Adult")
```

**Correct**

```python
if age == 18:
    print("Adult")
```

---

## 2. Indentation Errors

Python uses indentation to define code blocks.

**Incorrect**

```python
if age >= 18:
print("Adult")
```

**Correct**

```python
if age >= 18:
    print("Adult")
```

Proper indentation makes the code readable and prevents syntax errors.
