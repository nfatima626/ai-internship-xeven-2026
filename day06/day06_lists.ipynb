# 1. What is a List?

A **list** is an ordered and mutable collection in Python that allows us to store multiple values inside a single variable.

A list can store:

- Integers
- Floating-point numbers
- Strings
- Boolean values
- Other lists
- Mixed data types

```python

# 9. Adding Elements to a List

Python provides multiple methods to add new elements to a list.

The three most commonly used methods are:

- `append()`
- `insert()`
- `extend()`

Each method has a different purpose.

## append()

The `append()` method adds **one element** to the **end** of the list.

### Syntax

```python
list_name.append(value)
```

### Key Points

- Adds only one item.
- Always inserts at the end.
- Modifies the original list.

````python
students = ["Ali", "Sara", "Ahmed"]

students.append("Noor")

print(students)
````

### Example: Appending Different Data Types

````python
data = [10, 20]

data.append(30)
data.append("Python")
data.append(True)

print(data)
````

### Important Note

If you append another list, the entire list becomes **one single element**.

````python
numbers = [1, 2, 3]

numbers.append([4, 5])

print(numbers)
````

Output:

```text
[1, 2, 3, [4, 5]]
```

Notice that `[4, 5]` is added as **one element**, not two separate elements.

Output:

```text
[1, 2, 3, [4, 5]]
```

Notice that `[4, 5]` is added as **one element**, not two separate elements.

````python
students = ["Ali", "Ahmed"]

students.insert(1, "Sara")

print(students)
````

### Insert at the Beginning

````python
numbers = [20, 30, 40]

numbers.insert(0, 10)

print(numbers)
````

### Insert at the End

````python
numbers = [10, 20]

numbers.insert(len(numbers), 30)

print(numbers)
````

## extend()

The `extend()` method adds multiple elements from another iterable.

### Syntax

```python
list_name.extend(iterable)
```

### Key Points

- Adds multiple elements.
- Elements are added individually.
- Modifies the original list.

````python
students = ["Ali", "Sara"]

students.extend(["Ahmed", "Noor"])

print(students)
````

### Extend with Another List

````python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1.extend(list2)

print(list1)
````

## append() vs extend()


````python
numbers = [1, 2, 3]

numbers.append([4, 5])

print(numbers)
````

````python
numbers = [1, 2, 3]

numbers.extend([4, 5])

print(numbers)
````

### Difference

Using `append()`:

```text
[1, 2, 3, [4, 5]]
```

Using `extend()`:

```text
[1, 2, 3, 4, 5]
```

`append()` adds **one object**.

`extend()` adds **each element separately**.

# 10. Removing Elements

Python provides several methods to remove elements.

- `remove()`
- `pop()`
- `clear()`

## remove()

The `remove()` method removes the **first occurrence** of a value.

### Syntax

```python
list_name.remove(value)
```

````python
students = ["Ali", "Sara", "Ahmed"]

students.remove("Sara")

print(students)
````

### Removing Duplicate Values

````python
numbers = [10, 20, 20, 30]

numbers.remove(20)

print(numbers)
````

Only the **first occurrence** is removed.

## pop()

The `pop()` method removes an element using its index.

If no index is given, it removes the last element.

### Syntax

```python
list_name.pop(index)
```

````python
students = ["Ali", "Sara", "Ahmed"]

students.pop()

print(students)
````

````python
students = ["Ali", "Sara", "Ahmed"]

students.pop(1)

print(students)
````

### Saving the Removed Element

````python
numbers = [10, 20, 30]

removed = numbers.pop()

print("Removed:", removed)

print(numbers)
````

Unlike `remove()`, the `pop()` method **returns** the removed value.

## clear()

The `clear()` method removes all elements from the list.

### Syntax

```python
list_name.clear()
```

````python
students = ["Ali", "Sara", "Ahmed"]

students.clear()

print(students)
````

Output

```text
[]

A **nested list** is a list that contains one or more lists as its elements.

In simple words, it is a **list inside another list**.

Nested lists are useful when we need to store data in the form of **rows and columns**, such as:

- Student records
- Marksheets
- Tables
- Matrices
- Game boards
- Seating arrangements

## Syntax

```python

    [item1, item2],
    [item3, item4]

The first index selects the row.

The second index selects the column (element inside that row).

```

```python

Step 1

```

```

↓

```

```

Step 2

Another Example

```python

- First index → Select the row.

# 14. Copy vs Reference

When working with lists, it is important to understand the difference between **Copy** and **Reference**.

## Reference

When we assign one list to another variable using the assignment operator (`=`), **both variables point to the same list in memory**.

```
Reference

a ─────┐
        ▼
     [1, 2, 3]
        ▲
b ──────┘
```

Since both variables refer to the same object, modifying one variable also changes the other.

### Example

```python
a = [1, 2, 3]

b = a

b.append(4)

print(a)
print(b)
```

Output

```text
[1, 2, 3, 4]
[1, 2, 3, 4]
```

---

## Copy

If we want an independent list, we use the `copy()` method.

```
Copy

a ─────► [1, 2, 3]

b ─────► [1, 2, 3]
```

Now both variables refer to different list objects.

Changing one list does not affect the other.

### Example

```python
a = [1, 2, 3]

b = a.copy()

b.append(4)

print(a)
print(b)
```

Output

```text
[1, 2, 3]
[1, 2, 3, 4]
```

## Key Difference

| Reference (`=`) | Copy (`copy()`) |
|-----------------|-----------------|
| Same object | Different objects |
| Changes affect both variables | Changes affect only one list |
| No new memory is created | A new list is created |

````python
a = [1, 2, 3]

b = a

b.append(4)

print("a =", a)
print("b =", b)
````

````python
a = [1, 2, 3]

b = a.copy()

b.append(4)

print("a =", a)
print("b =", b)
````

# 15. List Comprehension

List Comprehension is a shorter and more Pythonic way to create a **new list** using a loop.

Instead of writing:

1. Create an empty list.
2. Write a `for` loop.
3. Append values one by one.

Python allows us to do all of this in a single line.

## Why Does List Comprehension Exist?

Many programmers repeatedly wrote code like this:

```python
result = []

for i in range(5):
    result.append(i)
```

Since this pattern is very common, Python introduced **List Comprehension** as a cleaner and more readable way to create new lists.

## Syntax

```python
[expression for variable in iterable]
```

## Traditional for Loop

```python
squares = []

for i in range(5):
    squares.append(i * i)

print(squares)
```

## List Comprehension

```python
squares = [i * i for i in range(5)]

print(squares)
```

Both programs produce the same output.

```text
[0, 1, 4, 9, 16]
```

---

## List Comprehension with Conditions

We can also filter elements while creating a list.

### Example: Even Numbers

```python
evens = [i for i in range(10) if i % 2 == 0]

print(evens)
```

Output

```text
[0, 2, 4, 6, 8]
```

---

### Example: Convert Names to Uppercase

```python
names = ["Ali", "Sara", "Ahmed"]

upper_names = [name.upper() for name in names]

print(upper_names)
```

Output

```text
['ALI', 'SARA', 'AHMED']
```

## Comparison

| Traditional Loop | List Comprehension |
|------------------|-------------------|
| More lines of code | Fewer lines of code |
| Uses `append()` | Creates the list directly |
| Good for complex logic | Best for simple list creation |

## When Should We Use List Comprehension?

When creating a **new list** from an existing iterable.

Examples:

- Squares
- Cubes
- Uppercase names
- Even numbers
- Odd numbers

## When Should We Avoid It?

Avoid list comprehension if the logic becomes too complex.

In such cases, a normal `for` loop is easier to read and maintain.

## Key Takeaways

- Creates a new list.
- Shorter than a traditional `for` loop.
- Improves readability for simple operations.
- Supports conditions using `if`.
- Does **not** replace every `for` loop.

````python
squares = []

for i in range(5):
    squares.append(i * i)

print(squares)
````

````python
squares = [i * i for i in range(5)]

print(squares)
````

````python
evens = [i for i in range(10) if i % 2 == 0]

print(evens)
````

````python
names = ["Ali", "Sara", "Ahmed"]

upper_names = [name.upper() for name in names]

print(upper_names)
````
