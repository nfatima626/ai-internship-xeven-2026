```python
print("=" * 40)
print("Arithmetic Operators")
print("=" * 40)

a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

```python
print("=" * 40)
print("Comparison Operators")
print("=" * 40)

x = 20
y = 15

print("Equal:", x == y)
print("Not Equal:", x != y)
print("Greater:", x > y)
print("Less:", x < y)
print("Greater or Equal:", x >= y)
print("Less or Equal:", x <= y)
```

```python
print("=" * 40)
print("Logical Operators")
print("=" * 40)

age = 20
has_id = True

print("AND:", age >= 18 and has_id)
print("OR:", age >= 18 or has_id)
print("NOT:", not has_id)
```

```python
print("=" * 40)
print("Type Conversion")
print("=" * 40)

num = "25"

print(int(num))
print(float(num))
print(str(100))
print(bool(1))
print(bool(0))
```

```python
print("=" * 40)
print("Operator Precedence")
print("=" * 40)

print(2 + 3 * 4)
print((2 + 3) * 4)

print(10 - 2 ** 3)
print((10 - 2) ** 3)
```

```python
print("=" * 40)
print("Type Conversion Error Handling")
print("=" * 40)

try:
    age = int(input("Enter age: "))
    print("Age:", age)

except ValueError:
    print("Invalid input. Please enter numbers only.")
```

## Article Summary

**Article:** Type Conversion in Python (GeeksforGeeks)

### Key Points

- The article explains that type conversion is the process of changing one data type into another.
- Python supports two types of type conversion: implicit (automatic) and explicit (manual).
- Explicit type conversion is performed using built-in functions such as `int()`, `float()`, `str()`, and `bool()`.
- Type conversion helps ensure that mathematical operations and data processing work correctly.
- The article recommends converting user input carefully and validating data before performing operations to avoid runtime errors.
- Understanding type conversion improves code reliability, readability, and reduces programming errors.
